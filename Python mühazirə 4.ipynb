```python
# indi isə daxilinə verdiyimiz datanı əlimizdə olan şərtlərə uyğun olaraq təmizləyən bir funksiya yazaq

"""
Məsələn, qiymət mənfi ola bilməz və ya 0 ola bilməz.

Gəlin istənilən sayda qiyməti qəbul edib, içindən yalnız yararlı olanları seçib qaytaran funksiya yazaq.
""" 

def temiz_data_sec(limit, *qiymetler):
    temiz_siyahi = [] # Boş qutu yaradırıq
    
    for x in qiymetler:
        # Şərt: Qiymət limitdən böyük olmalıdır (məsələn, 0-dan böyük)
        if x > limit:
            temiz_siyahi.append(x) # Qutuya atırıq
            
    return temiz_siyahi

# Funksiyanı çağırırıq: "Mənə 0-dan böyük olanları qaytar"
real_data = temiz_data_sec(0, 100, -50, 250, 0, -10, 500)

print("Təmizlənmiş data: {}".format(real_data))
# Nəticə: [100, 250, 500]
```

```python
# Adların Toqquşması (Kölgələmə) 

# hal hazirda biz localda olan deyiseni oradan cixarib globalda nece istifade ede bileceyimizi bilirik 
# bes global da olan deyiseni localin daxilinde deyismek ve ya istifade etmek istesek bu zaman ne etmeliyik ? 

xal = 0  # Oyunun başlanğıc xalı (Qlobal)

def xal_qazan():
    global xal  # Python-a deyirik: "Yeni xal yaratma, çöldəkini istifadə et!"
    xal = xal + 10

xal_qazan()
print(xal)  # Nəticə: 10 (Artıq çöldəki dəyişdi)
```

```python
# indi ise * arg ve global olan numune bir sade kod yazaq

# --- QLOBAL HİSSƏ (Konfiqurasiya) ---
# Bu dəyişən bütün funksiyalar üçün eyni qalır
MEZENNE = 1.70  # 1 USD = 1.70 AZN

def azn_cevir(*dollar_qiymetleri):
    """
    Verilən dollar qiymətlərini qlobal məzənnəyə vuraraq
    AZN siyahısı qaytarır.
    """
    yeni_siyahi = [] # Boş qutu
    
    # Gələn limitsiz sayda qiyməti tək-tək götürürük
    for dollar in dollar_qiymetleri:
        # Burada QLOBAL dəyişəndən (MEZENNE) istifadə edirik
        manat = dollar * MEZENNE  
        yeni_siyahi.append(manat)
        
    return yeni_siyahi

# --- İSTİFADƏ ---

# Ssenari 1: 3 fərqli məhsul satılıb
hesabat1 = azn_cevir(100, 50, 20)
print(f"Hesabat 1 (AZN): {hesabat1}")
# Nəticə: [170.0, 85.0, 34.0]

# Ssenari 2: Cəmi 1 böyük satış olub
hesabat2 = azn_cevir(1000)
print(f"Hesabat 2 (AZN): {hesabat2}")
# Nəticə: [1700.0]
```

```python
# raise 

def tersine_cevir(yazi):
    # 1. Yoxlayırıq: Gələn məlumat string (yazi) deyilmi?
    if (type(yazi) != str):
        # 2. Əgər string deyilsə, süni şəkildə XƏTA YARADIRIQ!
        raise ValueError("Zəhmət olmasa düzgün məlumat (yazı) daxil edin.")
    else:
        # 3. Əgər hər şey qaydasındadırsa, tərsinə çevir
        return yazi[::-1]

# --- YOXLAYAQ ---

# Ssenari 1: Düzgün istifadə
print(tersine_cevir("Python"))
# Nəticə: nohtyP

# Ssenari 2: Qayda pozuntusu (Rəqəm göndəririk)
print(tersine_cevir(12))
```

```python
"""
Tapşırıq: Neçə Taksi Lazımdır?
Ssenari: Siz "Bolt" və ya "Uber" üçün sadə bir alqoritm yazırsınız.

Bir taksi maksimum 4 nəfər tutur.

Turistlərin sayı daxil edilir.

Siz tapmalısınız ki, bu qədər adamı daşımaq üçün minimum neçə taksi lazımdır?
"""
```

```python
import math # Kitabxananı çağırırıq

def taksi_hesabla(adam_sayi):
    try:
        # 1. Məlumatı rəqəmə çevirməyə çalışırıq
        adam_sayi = int(adam_sayi)
        
        # 2. Mənfi adam ola bilməz (Kiçik yoxlama)
        if adam_sayi <= 0:
            print("Adam sayı müsbət olmalıdır!")
            return # Funksiyanı dayandırır

        # 3. Hesablama (Ən vacib hissə)
        # 1 taksi = 4 nəfər
        # Məsələn: 5 nəfər / 4 = 1.25 -> ceil -> 2 taksi
        taksi_sayi = math.ceil(adam_sayi / 4)
        
        print("{} nəfər üçün {} ədəd taksi lazımdır.".format(adam_sayi,taksi_sayi))
        
    except ValueError:
        # 4. Hərf yazılarsa
        print("XƏTA: Zəhmət olmasa adam sayını rəqəmlə yazın (Məs: 5, 10)")

taksi_hesabla(4)
```

```python
"""
Tapşırıq: "Dondurma Qonaqlığı"
Ssenari: Siz dostlarınıza dondurma almaq istəyirsiniz.

Cibinizdə müəyyən qədər pul var (Input 1).

Dondurmanın qiyməti bəllidir (Input 2).

Sual: Bu pula maksimum neçə dənə dondurma ala bilərsiniz
""" 
```

```python
import math

try:
    pul = float(input("ne qeder pulunuz var"))
    qiymet = float(input("dondurma neciyedir?"))

    if pul<=0 or qiymet<=0:
        print("pul ve ya qiymet 0-dan boyuk olmalidir")
    
    say = math.floor(pul/qiymet)
    qaliq = pul - (qiymet*say)

    print("sizin {} manat pulunuza {} ədəd dondurma düşür".format(pul,qiymet))
    print("size qaytarilacaq olan pul {} manatdir ".format(qaliq))
except ValueError:
    print("qiymet ve ya pul reqem olmalidir")
```
