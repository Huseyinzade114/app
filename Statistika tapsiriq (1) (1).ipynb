```python
import numpy as np

# 1. 5 fənnin siyahısını müəyyən edirik
fennler = ['Riyaziyyat', 'Fizika', 'Kimya', 'İnformatika', 'İngilis dili']

# 2. Hər sətir (tələbə) üzrə ortalamanı hesablayırıq
df['Ortalama'] = df[fennler].mean(axis=1).round(2)

# 3. Şərtə görə "Keçdi" və ya "Qaldı" statusunu veririk
# Əgər ortalama 70-dən böyükdürsə 'Keçdi', əks halda 'Qaldı'
df['Status'] = np.where(df['Ortalama'] > 70, 'Keçdi', 'Qaldı')

# Nəticəni görək
print(df[['Ad Soyad', 'Ortalama', 'Status']].head(10))
```

```python
# Fənn sütunlarını seçirik
fennler = ['Riyaziyyat', 'Fizika', 'Kimya', 'İnformatika', 'İngilis dili']

# 0.25 (Q1), 0.50 (Q2/Median) və 0.75 (Q3) kvartillərini hesablayırıq
kvartiller = df[fennler].quantile([0.25, 0.5, 0.75])

# Daha yaxşı oxunması üçün sətir adlarını dəyişirik
kvartiller.index = ['Q1 (25%)', 'Q2 (Median)', 'Q3 (75%)']

print(kvartiller)
```

```python
# 90% tələbənin hansı baldan aşağı qaldığını tapırıq
p90 = df[fennler].quantile(0.9)

print("Hər fənn üzrə 90-cı persentil (90% tələbə bu baldan aşağıdır):")
print(p90)
```

```python
fennler = ['Riyaziyyat', 'Fizika', 'Kimya', 'İnformatika', 'İngilis dili']

# 1. Q1 (25%) və Q3 (75%) tapırıq
Q1 = df[fennler].quantile(0.25)
Q3 = df[fennler].quantile(0.75)

# 2. IQR-ı hesablayırıq
IQR = Q3 - Q1

# 3. Min və Max sərhədləri (Whisker-lər) təyin edirik
min_serhed = Q1 - 1.5 * IQR
max_serhed = Q3 + 1.5 * IQR

# Nəticələri bir cədvəldə görək
limitler = pd.DataFrame({
    'Q1': Q1,
    'Q3': Q3,
    'IQR': IQR,
    'Alt Sərhəd (Min)': min_serhed,
    'Üst Sərhəd (Max)': max_serhed
})

print("Statistik Sərhədlər:")
print(limitler)
```

```python
# Fənn sütunlarını seçirik
fennler = ['Riyaziyyat', 'Fizika', 'Kimya', 'İnformatika', 'İngilis dili']

# 0.25 (Q1), 0.50 (Q2/Median) və 0.75 (Q3) kvartillərini hesablayırıq
kvartiller = df[fennler].quantile([0.25, 0.5, 0.75])

# Daha yaxşı oxunması üçün sətir adlarını dəyişirik
kvartiller.index = ['Q1 (25%)', 'Q2 (Median)', 'Q3 (75%)']

print(kvartiller)
```

```python
# Hər fənn üzrə outlier-ləri tapırıq
for fenn in fennler:
    
    outliers = df[
        (df[fenn] < min_serhed[fenn]) |
        (df[fenn] > max_serhed[fenn])
    ]
    
    print(f"\n------ {fenn} Outlierləri ------")
    print(outliers[[fenn]])
```

```python
# 1. Hesablama aparılacaq fənn sütunlarını seçirik
fennler = ['Riyaziyyat', 'Fizika', 'Kimya', 'İnformatika', 'İngilis dili']

# 2. .std() metodunu tətbiq edirik
stdev_neticeler = df[fennler].std().round(2)

print("Hər fənn üzrə Standart Meyil:")
print(stdev_neticeler)
```
