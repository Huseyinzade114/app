```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\ACER\Desktop\student_data.xlsx")
df
```

```python
# 1. Bolge sütunundakı null dəyərləri ən çox təkrarlanan (mode) dəyərlə dolduraq
# mode() funksiyası siyahı qaytardığı üçün [0] indeksini götürürük
most_frequent_bolge = df['Bolge'].mode()[0]
df['Bolge'] = df['Bolge'].fillna(most_frequent_bolge)

# 2. Miqdar sütunundakı null dəyərləri ədədi orta (mean) ilə dolduraq
average_miqdar = df['Miqdar'].mean()
df['Miqdar'] = df['Miqdar'].fillna(average_miqdar)

# 3. Endirim_% sütunundakı null dəyərləri 0 ilə dolduraq
df['Endirim_%'] = df['Endirim_%'].fillna(0)

# Nəticəni yoxlayaq
print(df.isnull().sum()) # Null dəyərlərin qalıb-qalmadığını görmək üçün
```

```python
# Vahid_Qiymet sütunundakı null dəyərləri median ilə dolduraq
median_qiymet = df['Vahid_Qiymet'].median()
df['Vahid_Qiymet'] = df['Vahid_Qiymet'].fillna(median_qiymet)

# Nəticəni yoxlayaq
print(df.isnull().sum()) # Null dəyərlərin qalıb-qalmadığını görmək üçün
```

```python
# 1. Sadəcə Hasil (Endirimsiz Məbləğ)
df['Umumi_Mebleg'] = df['Miqdar'] * df['Vahid_Qiymet']

# 2. Endirimi nəzərə alan Net Gəlir (Əsl qazanc)
df['Net_Gelir'] = df['Miqdar'] * df['Vahid_Qiymet'] * (1 - df['Endirim_%'] / 100)

# Nəticənin ilk 5 sətrinə baxaq
print(df.head(5))
```

```python
# NET GELIR HISSESINI 2 EDED QALMAQ SERTI ILE YUVARLAQLASDIRAQ
df['Net_Gelir'] = df['Net_Gelir'].round(2)
df.head(5)
```

```python
# 1. Bölgə üzrə Net Gəlir
bolge_pivot = df.pivot_table(index='Bolge', values='Net_Gelir', aggfunc='sum')

# 2. Kateqoriya üzrə Net Gəlir
kateqoriya_pivot = df.pivot_table(index='Kateqoriya', values='Net_Gelir', aggfunc='sum')

# 3. Məhsul üzrə Net Gəlir
mehsul_pivot = df.pivot_table(index='Mehsul', values='Net_Gelir', aggfunc='sum')

# Nəticələri ekrana çıxaraq
print("--- Bölgə üzrə Gəlir ---")
print(bolge_pivot)
print("\n--- Kateqoriya üzrə Gəlir ---")
print(kateqoriya_pivot)
print("\n--- Məhsul üzrə Gəlir ---")
print(mehsul_pivot)
```

```python
# Kateqoriya sətirlərdə, Məhsullar sütunlarda olmaqla net gəlir cəmi
kat_mehsul_pivot = df.pivot_table(index='Kateqoriya', 
                                  columns='Bolge', 
                                  values='Net_Gelir', 
                                  aggfunc='sum')



# Nəticəni göstərək
print(kat_mehsul_pivot)
```

```python
# Null (NaN) dəyərləri 0 ilə əvəzləyək ki, daha təmiz görünsün
# (Çünki bir kateqoriyaya aid olmayan məhsul orada NaN görsənəcək)
kat_mehsul_pivot = kat_mehsul_pivot.fillna(0)
kat_mehsul_pivot
```
