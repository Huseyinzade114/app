```python
"""
List — bir neçə elementi tək dəyişəndə saxlamağa imkan verən düzülmüş (ordered) və dəyişdirilə bilən (mutable) verilənlər strukturudur

Düzülmüşdür → elementlərin sırası dəyişməz (siz dəyişdirməyənə qədər).
Mutable-dır → istədiyiniz elementi dəyişə, silə, əlavə edə bilərsiniz.
Müxtəlif tip element saxlaya bilər → int, float, string, hətta başqa list.
İndekslənir → sıfırdan başlayaraq elementlərə indekslə daxil olunur.
"""

liste =  [3,4,5,6,"jet",3.14,5.324]
print(liste)
print(type(liste))
```

```python
"""
list() funksiyası
list() funksiyası istənilən iterable (dəyərlər ardıcıllığı) obyektini list tipinə çevirir
iterable → string, tuple, set, dictionary, range və s.
"""

#String → List 
a = "salam"
print(a)
print(type(a))
b = list(a)
print(b)
print(type(b))
```

```python
# Aşağıda Python list data tipində indeksləmə və parçalama (indexing & slicing) çox sadə və qısa şəkildə izah olunur.
# İndeks — siyahıda elementlərin sıra nömrəsidir. 

my_list = ["vahid", 2, 2.4 , "jet"]

print(my_list[0])  # vahid
print(my_list[1])  # 2
print(my_list[3],end="\n\n")  # jet

# Mənfi indeksləmə (sondan sayır) 
print(my_list[-1])  # jet
print(my_list[-2])  # 2.4
print(my_list[-4])  # vahid

```

```python
# Parçalama (Slicing)
# Slicing siyahının bir hissəsini götürmək üçündür:
# start → haradan başlasın
# end → harada bitsin (bu indeks daxil deyil)
# step → neçə-neçə keçsin

numbers = [10, 20, 30, 40, 50, 60]
print(" başlanğıc və son dəyər verilib")
print(numbers[1:4])  
# [20, 30, 40]  → 1-dən başlayır, 4 daxil deyil

# Başlanğıc yazılmırsa → 0-dan başlayır
print(" başlanğıc yazılmır ama son dəyər verilib")
print(numbers[:3])
# [10, 20, 30]

# Son yazılmırsa → sona qədər götürür
print(" başlanğıc verilib ama son dəyər verilməyib")
print(numbers[3:])
# [40, 50, 60]

# Step ilə parçalama
print(" step parçalama ilə verilmiş ")
print(numbers[0:6:2])
# [10, 30, 50] → hər 2 elementi götürür
```

```python
"""
Tuple (demət) dəyişdirilə bilməyən (immutable) olduğu üçün onun metodları çox azdır.
Yəni append(), remove(), pop() kimi metodlar tuple-da yoxdur.

Tuple üçün əsas istifadə olunan iki metod var:
""" 

# count() — dəyərin neçə dəfə təkrarlandığını sayır 

t = (1, 2, 2, 3, 2)

print(t.count(2))  
# Nəticə: 3   (tuple-da "2" üç dəfə var)
```

```python
"""
Dictionary — Python-da anahtar–dəyər (key–value) cütlərini saxlayan xüsusi bir data tipidir.
List və tuple-dan fərqli olaraq dictionary-də indeks yoxdur.
Hər bir element belə saxlanılır:
key : value
"""
# Dictionary necə yaradılır?

student = {
    "name": "Aysel",
    "age": 20,
    "grades": [90, 85, 92]
}

# Dictionary-də dəyərə necə müraciət edirik? 
print(student,end="\n\n")
print(student["name"])   # Aysel
print(student["age"])    # 20

```

```python
# indi ise dictionary data tipi ile islenen bezi metodlara baxaq bu dersde sadece 2 metoda baxaciq 
# gorduyumuz kimi dictionary veri tipinde key values deyerleri var onlari ayrica almaq ucun metodlardan istifade etmek olar

# keys() – Bütün anahtarları (key-ləri) qaytarır
d = {"name": "Aysel", "age": 20}

print(d.keys())
# dict_keys(['name', 'age'])

```

```python
# values() – Bütün dəyərləri qaytarır

print(d.values())
# dict_values(['Aysel', 20])
```

```python
"""
in operatoru - 
Python-da in operatoru bir elementin hər hansı bir siyahıda, demətdə (tuple), sözlükdə (dictionary key-lərində) və ya sətirdə (string) mövcud olub-olmadığını yoxlamaq 
üçün istifadə olunur.
Əgər element həmin kolleksiyada varsa → True,
əks halda → False qaytarır.
""" 
numbers = [1, 2, 3, 4]
print(3 in numbers)   # True
print(10 in numbers)  # False

t = ("Aysel", "Zaur", "Nigar")
print("Zaur" in t)    # True
print("Elmar" in t)   # False

s = "Baku"
print("a" in s)   # True
print("z" in s)   # False

# Dictionary-də in yalnız key-ləri yoxlayır 
student = {"name": "Aysel", "age": 20}
print("name" in student)   # True
print("Aysel" in student)  # False  (çünki dəyərlərinə baxmır!)


```

```python
students = {
    "Aysel": 95,
    "Murad": 82,
    "Kamran": 76,
    "Zaur": 65,
    "Nigar": 48
}

for name, score in students.items():
    if score >= 90:
        grade = "A"
        status = "Keçdi"
        
    elif score >= 80:
        grade = "B"
        status = "Keçdi"
    elif score >= 70:
        grade = "C"
        status = "Keçdi"
    elif score >= 60:
        grade = "D"
        status = "Şərti keçdi"
    else:
        grade = "F"
        status = "Kecmədi"

    print(name, "-", score, "→ Qiymət:", grade, "| Status:", status)
```

```python
# infinite loop 

"""
sonsuz döngü deməkdir, burada diqqətli olmalıyıq 
şərti verərkən mütləq incələmək lazımdır müəyyən bir yerdən sonra false dəyər almalıdır əks təqdirdə proqram sonsuz olaraq çalışar və buda
bizim istəmiyəcəyimiz birşeydir
yuxarılarda qeyd etdiyimiz nümunələrin birinə geri dönək və bu məsələni başa düşməyə çalışaq.

i += 1   # i = i + 1 bu kodda əgər qeyd etdiyimiz sətri yazmazsaq nə baş verər? 
i həmişə 1 olaraq qalar yəni i dəyərinin üzərinə 1 gəlməz bu zaman artıq i həmişə 5 dən kiçik olar və while həmişə true olduğu üçün 
kod durmadan çalışar
"""

i = 1

while i <= 5:
    print(i)
    i += 1   # i = i + 1


```

```python
# while ve for birlikte islense nece besssss?

grades = []  #60,70     # bal siyahısı
i = 0 #2             # sayğac

while i < 5:      # 5 bal daxil ediləcək
    score = int(input("Balı daxil edin: "))
    grades.append(score)
    i = i + 1     # sayğacı artırırıq

print("Nəticələr:")
for bal in grades:
    if bal >= 60:
        print(bal, "- Keçdi")
    else:
        print(bal, "- Kecmədi")
```

```python
dogru_sifre = "1234"
sansa_sayi = 0

while sansa_sayi < 3:
    sifre = input("Şifrəni daxil edin: ")
    
    if sifre == dogru_sifre:
        print("Giriş uğurludur!")
        sansa_sayi = 3      # Döngünü bitirmək üçün şansları doldururuq
    else:
        print("Yanlış şifrə!")
        sansa_sayi = sansa_sayi + 1

if sifre != dogru_sifre:
    print("3 cəhd bitdi. Giriş bloklandı!")
```

```python
# break və continue ifadələri 

# break işlədildiyi anda döngü tamamilə dayanır və döngüdən çıxılır.

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# Continue ifadesi break ifadesine gore daha az istifade olunur 
# continue işlədildiyi zaman həmin iterasiyadakı qalan əmrlər keçilir → döngü növbəti elementdən davam edir.

for i in range(6):
    if i == 3:
        continue
    print(i)
```

```python
mehsullar = ["çörək", "süd", "yumurta", "yağ"]

while True:
    axtar = input("Məhsul adı daxil edin (dayandırmaq üçün 'stop'): ")

    if axtar == "stop":
        print("Axtarış dayandırıldı.")
        break          # döngünü dayandırır

    if axtar not in mehsullar:
        print("Bu məhsul tapılmadı! → davam edirəm")
        continue       # siyahıda yoxdursa növbəti iterasiyaya keç
       
    print("Məhsul tapıldı:", axtar)
```
