```python
# Filter

data = {
    'Ad': ['Aygün', 'Elvin', 'Nigar', 'Samir', 'Orxan', 'Lalə'],
    'Sobesi': ['IT', 'HR', 'IT', 'Maliyyə', 'HR', 'IT'],
    'Maas': [1200, 800, 1500, 950, 750, 2000],
    'Seher': ['Bakı', 'Gəncə', 'Bakı', 'Sumqayıt', 'Bakı', 'Naxçıvan']
}

df = pd.DataFrame(data)
```

```python
# arxa planda nə baş verir?
# Siz df['Maas'] > 1000 yazanda Pandas sizə əslində datanı vermir, o sizə True (Doğru) və False (Yanlış) siyahısı verir.

# Sadəcə şərti yoxlayaq
print(df['Maas'] > 1000)
```

```python
# Biz bu True/False siyahısını DataFrame-in içinə yerləşdirəndə, Pandas yalnız True olan sətirləri bizə göstərir.

yuksek_maas = df[df['Maas'] > 1000]
print(yuksek_maas)
```

```python
# "Şöbəsi 'IT' olan və maaşı 1300-dən çox olanlar." 

# (Şərt 1) & (Şərt 2)
senior_it = df[(df['Sobesi'] == 'IT') & (df['Maas'] > 1300)]
senior_it
```

```python
# Ssenari: "Ya 'HR' şöbəsində işləyənlər, ya da maaşı 900-dən az olanlar."

# (Şərt 1) | (Şərt 2)
filter_or = df[(df['Sobesi'] == 'HR') | (df['Maas'] < 900)]
filter_or
```

```python
# (Şərt 1) & (Şərt 2)
senior_it = df[(df['Sobesi'] == 'IT') and (df['Maas'] > 1300)]
senior_it
```

```python
filter_or = df[(df['Sobesi'] == 'HR') or (df['Maas'] < 900)]
filter_or 
```

```python
# .isin() – Siyahı üzrə yoxlama

# Ssenari: "İşçilər 'Bakı' və ya 'Gəncə' və ya 'Naxçıvan' şəhərindəndirsə gətir." 
# Uzun yol (Pis yol): df[(df['Seher'] == 'Bakı') | (df['Seher'] == 'Gəncə') | ...]

# Qısa yol (Professional yol):
seherler = ['Bakı', 'Gəncə', 'Naxçıvan']
secilenler = df[df['Seher'].isin(seherler)]
secilenler
```

```python
# .between() – Aralıq yoxlama
# Ssenari: "Maaşı 800 ilə 1200 arasında olanlar."
# Uzun yol: df[(df['Maas'] >= 800) & (df['Maas'] <= 1200)]

orta_maas = df[df['Maas'].between(800, 1200)]
orta_maas
```

```python
# String filter
# Ssenari: "Adının içində 'a' hərfi olan hər kəsi tap." (SQL-dəki LIKE '%a%' kimi).

# str.contains() metodu
a_herfi_olanlar = df[df['Ad'].str.contains('a', case=False)] 
# case=False: Böyük-kiçik hərf fərqini iqnor edir
a_herfi_olanlar
```

```python
# df (Y vərəqi) üçün ad dəyişimi
df = df.rename(columns={'Faiz': 'Faiz_Y'})

# df2 (X vərəqi) üçün ad dəyişimi
df2 = df2.rename(columns={'Faiz': 'Faiz_X'})

# Yoxlayaq
print(df.head(2))
print(df2.head(2))
```

```python
# cedvelleri birleşdirme

# 1-ci Hissə: Əgər df və df2 (Sizin Exceldən oxuduqlarınız) hazırdırsa, əvvəlcə onları birləşdirək
# Bu sətir df_main-i yaradır:
df_main = pd.merge(df, df2, on='İl', how='inner')
df_main

```

```python
# Dörd Əsas Əməliyyat

# 1. Toplama (Cəm göstərici)
df_main['Cem_Faiz'] = df_main['Faiz_X'] + df_main['Faiz_Y']

# 2. Çıxma (Fərq)
df_main['Ferq'] = df_main['Faiz_X'] - df_main['Faiz_Y']

# 3. Vurma (Məsələn, faizi əmsala vurmaq)
df_main['Faiz_X_2qat'] = df_main['Faiz_X'] * 2

# 4. Mürəkkəb Hesablama (Ortalama)
# Mötərizədən istifadə etmək mütləqdir!
df_main['Ortalama'] = (df_main['Faiz_X'] + df_main['Faiz_Y']) / 2

print(df_main)
```

```python
# Statistik Əməliyyatlar (Sütun daxili)
# Bəzən bir sütunu digəri ilə yox, öz daxilində hesablamaq lazımdır.

# Bütün Faiz_X sütununu cəmləmək
total_x = df_main['Faiz_X'].sum()
print(total_x)

# Kümülatif Cəm (İllər üzrə yığılaraq gedən artım)
df2['Kumulyativ_X'] = df_main['Faiz_X'].cumsum()
print(df2)
```

```python
# pivot 

import pandas as pd
import numpy as np

data = {
    'Seher': ['Bakı', 'Bakı', 'Bakı', 'Gəncə', 'Gəncə', 'Sumqayıt', 'Sumqayıt'],
    'Mehsul': ['Telefon', 'TV', 'Telefon', 'TV', 'Telefon', 'Telefon', 'TV'],
    'Satis': [1000, 2000, 1200, 1500, 800, 900, 1800],
    'Miqdar': [2, 1, 3, 2, 4, 2, 1]
}

df = pd.DataFrame(data)
print("--- İlkin Data ---")
print(df)
```

```python
pivot1 = df.pivot_table(index='Seher', values='Satis', aggfunc='sum')
print(pivot1)
```

```python
pivot2 = df.pivot_table(index='Seher',      # Sətirlərdə Şəhər olsun
                        columns='Mehsul',   # Sütunlarda Məhsul olsun
                        values='Satis',     # Hesablanan Satış olsun
                        aggfunc='sum')      # Cəmləsin

print(pivot2)
```

```python
# Boşluqları Doldurmaq (fill_value)

# Məsələn, Gəncədə TV satışı yoxdursa, 0 yazsın
pivot2 = df.pivot_table(index='Seher', columns='Mehsul', values='Satis', aggfunc='sum', fill_value=0)
pivot2
```

```python
# Cəmi Göstərmək (margins)
# Excel-dəki "Grand Total" (Yekun Cəm) sətrini əlavə etmək üçün.

pivot_total = df.pivot_table(index='Seher', 
                             columns='Mehsul', 
                             values='Satis', 
                             aggfunc='sum',
                             margins=True,       # Yekun cəmi aktiv et
                             margins_name='CƏM') # Adını qoy

print(pivot_total)
```

```python
# Çoxlu Funksiya Tətbiqi
# Analitik eyni anda həm Satışın Cəmini, həm də Satılan Miqdarın Ortalamasını görmək istəyə bilər.

pivot_multi = df.pivot_table(index='Seher',
                             values=['Satis', 'Miqdar'], # İki fərqli sütun
                             aggfunc={'Satis': 'sum', 'Miqdar': 'mean'}) # Hərəsinə ayrı funksiya

print(pivot_multi)
```
