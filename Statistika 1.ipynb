```python
import pandas as pd
import numpy as np
```

```python
# 1. Excel faylını oxuyuruq
df = pd.read_excel(r'C:\Users\ACER\Desktop\Data statistika 1.xlsx')
df
```

```python
# Pandas ilə Ortalama (Mean) hesablamaq
# df['Maas'] - sadəcə maaş sütununu götürür
pandas_mean = df['Maas'].mean()
print(pandas_mean)
print(f"Pandas ilə hesablanan Orta Maaş: {pandas_mean:.2f} AZN")
```

```python
# NumPy ilə Ortalama hesablamaq
# Eyni nəticəni verir, riyazi yanaşmadır
numpy_mean = np.mean(df['Maas'])
print(numpy_mean)
print(f"Numpy ile hesablanan orta maas: {numpy_mean:.2f} AZN")
```

```python
# bes şöbeler üçün ayrı-ayrı ortalama tapmaq istəsək onda necə edirik? 
# yol 1 -  menfi terefi odurki şirketde şöbe sayı çox olarsa çox setir yazmaq lazım gelecek

# 1. Sadəcə IT şöbəsini seçirik
it_iscileri = df[df['Şöbə'] == 'IT']
it_ortalama = it_iscileri['Maas'].mean()

# 2. Sadəcə HR şöbəsini seçirik
hr_iscileri = df[df['Şöbə'] == 'HR']
hr_ortalama = hr_iscileri['Maas'].mean()

print(f"IT Şöbəsinin Ortalaması: {it_ortalama}")
print(f"HR Şöbəsinin Ortalaması: {hr_ortalama}")
```

```python
# 2 ci yol Pivot Table mentiqidir 
# pivot_table istifadə edərək
# values = Hesablayacağımız sütun (Maaş)
# index = Sətirlərdə nə olsun? (Şöbə)
# aggfunc = Hansı funksiya? ('mean' - ortalama)

pivot_netice = df.pivot_table(values='Maas', index='Şöbə', aggfunc='mean')

print("--- Pivot Table ilə Nəticə ---")
print(pivot_netice)
```

```python
# GroupBy mentiqi 
# Şöbələr üzrə qruplaşdırıb ortalamanı tapırıq
# Məntiq: df.groupby('HANSINA_GÖRƏ_QRUPLAŞDIRIM')['HANSI_SÜTUNU_HESABLAYIM'].mean()

sobe_ortalamasi = df.groupby('Şöbə')['Maas'].mean()
sobe_ortalamasi
```

```python
# 1. Pandas ilə Median
# Diqqət: Pandas arxada özü sıralayır, bizim "sort" etməyimizə ehtiyac qalmır.
median_pandas = df['Maas'].median()
median_pandas
```

```python
# NumPy ilə Median
median_numpy = np.median(df['Maas'])
median_numpy
```

```python
sobe_modasi = df['Şöbə'].mode()
sobe_modasi
```

```python
# 1. Excel faylını oxuyuruq
df = pd.read_excel(r'C:\Users\ACER\Desktop\Data statistika 1.xlsx',sheet_name="stdev")
df
```

```python
# Əvvəlcə 'Bal' sütununu df-dən çıxarıb 'ballar' adında bir dəyişənə qoyuruq
ballar = df['Bal'] 

# İndi artıq 'ballar' dəyişənini istifadə edə bilərik
ortalama = ballar.mean()
variance = ballar.var()
std_dev = ballar.std()

print(f"Ortalama: {ortalama}")
print(f"varyans: {variance}")
print(f"Standart Kənarlaşma: {std_dev}")
```
