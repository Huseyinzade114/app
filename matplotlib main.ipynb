```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python

np.random.seed(42)

data = {
    "Student": [f"Student_{i}" for i in range(1, 51)],
    "Study_Hours": np.random.randint(1, 10, 50),
    "Exam_Score": np.random.randint(40, 100, 50),
    "Attendance": np.random.randint(50, 100, 50),
    "Department": np.random.choice(["Finance", "IT", "Economics"], 50)
}

df = pd.DataFrame(data)

print(df.head(3))
```

```python
plt.figure(figsize=(8, 5))

plt.scatter(df["Study_Hours"], df["Exam_Score"])

plt.title("Study Hours və Exam Score arasında əlaqə")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")

plt.show()
```

```python
# `df["Study_Hours"]` sütununun `df["Exam_Score"]` sütunu ilə korrelyasiyası
a = df["Study_Hours"].corr(df["Exam_Score"])
a
```

```python
plt.figure(figsize=(10, 5))

plt.plot(df["Student"], df["Exam_Score"])

plt.title("Tələbələrin imtahan nəticələri")
plt.xlabel("Student")
plt.ylabel("Exam Score")

plt.xticks(rotation=90)

plt.show()
```

```python
department_scores = df.groupby("Department")["Exam_Score"].mean()
print(department_scores,end="\n\n")

plt.figure(figsize=(7, 5))

plt.bar(department_scores.index, department_scores.values)

plt.title("Departamentlər üzrə orta imtahan balı")
plt.xlabel("Department")
plt.ylabel("Average Exam Score")

plt.show()
```

```python
plt.figure(figsize=(8, 5))

plt.hist(df["Exam_Score"], bins=10)

plt.title("Exam Score paylanması")
plt.xlabel("Exam Score")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(7, 5))

plt.boxplot(df["Exam_Score"])

plt.title("Exam Score Boxplot")
plt.ylabel("Exam Score")

plt.show()
```

```python
finance_scores = df[df["Department"] == "Finance"]["Exam_Score"]
it_scores = df[df["Department"] == "IT"]["Exam_Score"]
economics_scores = df[df["Department"] == "Economics"]["Exam_Score"]

plt.figure(figsize=(8, 5))

plt.boxplot(
    [finance_scores, it_scores, economics_scores],
    labels=["Finance", "IT", "Economics"]
)

plt.title("Departamentlər üzrə Exam Score Boxplot")
plt.xlabel("Department")
plt.ylabel("Exam Score")

plt.show()
```

```python
department_counts = df["Department"].value_counts()
print(department_counts)

plt.figure(figsize=(6, 6))

plt.pie(
    department_counts.values,
    labels=department_counts.index,
    autopct="%1.1f%%"
)

plt.title("Departamentlər üzrə tələbə bölgüsü")

plt.show()
```

```python
numeric_df = df[["Study_Hours", "Exam_Score", "Attendance"]]

corr = numeric_df.corr()

print(corr)
```

```python
"""
Qrup sayı = k = 7
Minimum = datadakı ən kiçik dəyər
Maximum = datadakı ən böyük dəyər        #k    qrup genisliyi
Aralıq = Maximum - Minimum
Qrup genişliyi = Aralıq / k
"""


# k = 1 + 3.322 * log10(n)   
```

```python

import math

scores = df["Exam_Score"]

n = len(scores)
minimum = scores.min()
maximum = scores.max()

print("Müşahidə sayı:", n)
print("Minimum:", minimum)
print("Maksimum:", maximum)
```

```python
k = 1 + 3.322 * np.log10(n)

print("Qrup sayı:", k)
k = math.ceil(k)

print("Yuvarlaq qrup sayı:", k)
```

```python
interval_width = (maximum - minimum) / k

interval_width = math.ceil(interval_width)

print("Yuvarlaq interval genişliyi:", interval_width)
```
