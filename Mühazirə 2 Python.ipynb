```python
# kvadrat tənliyin həll edilməsinin avtomatlaşdırılması  ( dəyərləri istifadəçidən al float olaraq)
# ax2+bx+c=0
# D=b**2−4ac   x1 = -b + D**0.5 / 2a   x2 = -b - D**0.5 / 2a

""""
riyaziyyatdan bildiyimiz kimi diskriminant 0 a bərabər olarsa tənliyin bir kökü - olarsa kökü yoxdur + olarsa 2 kökü var ama bu tapşırıqda 
bu qaydanı istinad edecik çünki if elif bloklarını keçenden sonra bunu edecik
"""

a = float(input("a dəyərini qeyd edin:"))
b = float(input(" b dəyərini qeyd edin:"))
c = float(input(" c dəyətini qeyd edin:"))

D = b**2 - 4*a*c
x1 = -b + D**0.5 / 2*a 
x2 = -b - D**0.5 / 2*a 

print("qeyd etdiyiniz kvadrat tənliyin həlli aşağıdakı kimidir:")

print("diskriminant=",D)
print("tənliyin birinci kökü=",x1)
print("tənliyin ikinci kökü=",x2)
```

```python
# Kvadrat tənliyin həll edilməsi (a, b, c istifadəçidən float olaraq alınır)
# ax^2 + bx + c = 0
# a = 1, b = 5, c = 6 - iki kok
# a = 1, b = 4, c = 4 - bir kok

a = float(input("a dəyərini daxil edin: "))
b = float(input("b dəyərini daxil edin: "))
c = float(input("c dəyərini daxil edin: "))

# Diskriminant hesablama
D = b**2 - 4*a*c

print("\nDiskriminant (D) =", D)

# Şərtli hallar:
if D > 0:
    x1 = (-b + (D ** 0.5)) / (2 * a)
    x2 = (-b - (D ** 0.5)) / (2 * a)
    print("Tənliyin iki həqiqi kökü var:")
    print("x1 =", x1)
    print("x2 =", x2)

elif D == 0:
    x = -b / (2 * a)
    print("Tənliyin bir dənə (təkrarlanan) kökü var:")
    print("x =", x)

else:
    print("Diskriminant mənfidir, tənliyin həqiqi kökü yoxdur.")
```
