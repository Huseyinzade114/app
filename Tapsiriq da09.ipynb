```python
"""
Təsəvvür edin ki, siz bir mağazanın data analitikisiniz. 
Menecer sizdən daxil edilən satış məlumatlarına əsasən məhsulun mənfəətini (profit) hesablamağı və həmin m
əhsulun şirkət üçün nə dərəcədə gəlirli olduğunu müəyyən edən kiçik bir proqram yazmağınızı xahiş edir.

Mənfəəti hesablayın (Ümumi Məbləğ - Ümumi Xərc).
Mənfəət marjasını (faizlə) hesablayın: (Mənfəət / Ümumi Məbləğ) * 100

Marja 40%-dən çoxdursa: "Yüksək gəlirli məhsul"
Marja 15% ilə 40% (daxil) arasındadırsa: "Orta gəlirli məhsul"
Marja 0% ilə 15% arasındadırsa: "Aşağı gəlirli məhsul"
Əgər xərc gəlirdən çoxdursa və ya bərabərdirsə: "Zərər və ya Gəlirsiz"
"""   
```

```python
# 1. İstifadəçidən məlumatların alınması
mehsul_adi = input("Məhsulun adını daxil edin: ")
satis_meblegi = float(input("Satışdan gələn ümumi məbləği daxil edin (AZN): "))
xerc = float(input("Məhsulun ümumi xərcini daxil edin (AZN): "))

# Sıfıra bölmə xətasının qarşısının alınması
if satis_meblegi == 0:
    print("Xəta: Satış məbləği 0 ola bilməz. Zəhmət olmasa düzgün məlumat daxil edin.")
else:
    # 2 və 3. Riyazi hesablamalar
    menfeet = satis_meblegi - xerc
    marja = (menfeet / satis_meblegi) * 100

    # 4. Şərtlər və Kateqoriyalaşdırma
    if marja > 40:
        kateqoriya = "Yüksək gəlirli məhsul"
    elif marja >= 15:
        kateqoriya = "Orta gəlirli məhsul"
    elif marja > 0:
        kateqoriya = "Aşağı gəlirli məhsul"
    else:
        kateqoriya = "Zərər və ya Gəlirsiz"

    # 5. Nəticənin çap edilməsi
    print("--------------------------------------------------")
    print("Analizin Nəticəsi:")
    print("Məhsul: " + mehsul_adi)
    print("Xalis Mənfəət: " + str(menfeet) + " AZN")
    print("Kateqoriya: " + kateqoriya)
    print("--------------------------------------------------")
```

```python
""" 
Siz bir aqrar və ya sənaye obyektindəki ağıllı su çənlərinə nəzarət edən data analitiksiniz. 
Çənin üzərindəki sensorlar hər saat mərkəzi bazaya iki məlumat göndərir: 
Su səviyyəsi (faizlə) və Temperatur (Selsi ilə). Lakin bəzən sensorlar xarab olur və qeyri-məntiqi datalar 
(məsələn, mənfi su səviyyəsi) göndərir
yaxud da kritik vəziyyətlər yaranır. Sizin vəzifəniz daxil olan datanı 
yoxlayan və uyğun statusu müəyyən edən alqoritm yazmaqdır.

Qeyri-məntiqi data xətası: Əgər su səviyyəsi 0-dan kiçikdirsə YAXUD 100-dən böyükdürsə
sistem "XƏTA: Sensor sıradan çıxıb, qeyri-mümkün data!" xəbərdarlığını verməlidir.

Kritik Təhlükə: Əgər su səviyyəsi 15%-dən azdırsa VƏ temperatur 40 dərəcədən çoxdursa
sistem "KRİTİK: Su tükənir və aşırı isinmə var! Təcili müdaxilə!" xəbərdarlığını verməlidir.

Daşma Riski: Əgər su səviyyəsi 90%-dən böyükdürsə
sistem "XƏBƏRDARLIQ: Çən daşmaq üzrədir!" deməlidir.

Normal Vəziyyət: Yuxarıdakı heç bir şərt ödənmirsə
sistem "Sistem normal işləyir." mesajını qaytarmalıdır.
"""
```

```python
# 1. Dataların daxil edilməsi
sensor_kodu = input("Sensorun kodunu daxil edin (məs: SN-001): ")
su_seviyyesi = float(input("Suyun səviyyəsini daxil edin (%): "))
temperatur = float(input("Suyun temperaturunu daxil edin (°C): "))

# 2. Şərtlər və Məntiqi Operatorlarla (and / or) Yoxlama
if su_seviyyesi < 0 or su_seviyyesi > 100:
    status = "XƏTA: Sensor sıradan çıxıb, qeyri-mümkün data!"
elif su_seviyyesi < 15 and temperatur > 40:
    status = "KRİTİK: Su tükənir və aşırı istinmə var! Təcili müdaxilə!"
elif su_seviyyesi > 90:
    status = "XƏBƏRDARLIQ: Çən daşmaq üzrədir!"
else:
    status = "Sistem normal işləyir."

# 3. Nəticənin çap edilməsi
print("\n--- Sensor Monitorinq Nəticəsi ---")
print("Sensor ID: " + sensor_kodu)
print("Sistemin Statusu: " + status)
print("----------------------------------")
```
